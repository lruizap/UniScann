```python
# Importar librerías necesarias
import os
import yaml
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import numpy as np
from pathlib import Path
import torch
```

```python
# Verificar si CUDA está disponible
print(f"CUDA disponible: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
```

```python
# Cargar y verificar el archivo data.yaml
data_path = "data.yaml"  # Ajusta la ruta a tu archivo data.yaml

# Verificar que el archivo existe
if os.path.exists(data_path):
    with open(data_path, 'r') as file:
        data_config = yaml.safe_load(file)

    print("Configuración del dataset:")
    print(
        f"Path de entrenamiento: {data_config.get('train', 'No especificado')}")
    print(f"Path de validación: {data_config.get('val', 'No especificado')}")
    print(f"Número de clases: {data_config.get('nc', 'No especificado')}")
    print(f"Nombres de clases: {data_config.get('names', 'No especificado')}")
else:
    print("¡Error! No se encontró el archivo data.yaml")
    print("Asegúrate de que el archivo esté en la ruta correcta")
```
